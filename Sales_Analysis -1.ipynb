Q-1 Which product category is more popular or sold among different genders?

```python
# Product Category vs Amount represented in lineplot for different genders

sns.set(rc={'figure.figsize':(30,5)})
sns.set_theme(style="ticks")
sns.lineplot(data=sales_1S, x= 'Product_Category', y= 'Amount' , hue = 'Gender')
```

```python
#line Plot for Product Category vs Amount for different Gender
sns.set(rc={'figure.figsize':(25,5)})
sns.set_theme(style="ticks")
plt.subplot(1,2,1)
sns.lineplot(data=sales_1S.loc[sales_1S.Gender=='M'], x= 'Product_Category', y= 'Amount', hue = 'Gender')
plt.subplot(1,2,2)
sns.lineplot(data=sales_1S.loc[sales_1S.Gender=='F'], x= 'Product_Category', y= 'Amount', hue = 'Gender')
```

Q-3 what category married people are buying?

```python
df
```

```python
Married=df.groupby(['Marital_Status','Product_Category'])['Amount'].sum().reset_index()
Married
```

```python
# Product Category vs Amount represented in lineplot for married or unmmaried people

sns.set(rc={'figure.figsize':(30,5)})
sns.set_theme(style="ticks")
sns.lineplot(data=Married,x='Product_Category', y='Amount', hue='Marital_Status')
```

```python
#Bar Plot for Product Category vs Amount for married or unmmaried people
sns.set(rc={'figure.figsize':(25,5)})
sns.set_theme(style="ticks")
plt.subplot(1,2,1)
sns.lineplot(data=Married.loc[Married.Marital_Status==0], x= 'Product_Category', y= 'Amount', hue = 'Marital_Status')

plt.subplot(1,2,2)
sns.lineplot(data=Married.loc[Married.Marital_Status==1], x= 'Product_Category', y= 'Amount', hue = 'Marital_Status')
```
